```
# Notice: do not change these function name


#Check if the number is a palindrome or not
def is_palindrome(x):
    # add your code here
    if(x<0):
        return False
    org=x
    rev=0
    while(x>0):
        rev=rev*10+(x%10)
        x=x//10
    if(rev!=org):
        return False
    else: 
        return True 

#Check if one string is an anagram of the other    
def is_anagrams(s, t):
    # add your code here
    d1={}
    if len(s) != len(t):
        return False;
    
    for x in s:
        if x not in d1:
            d1[x]=1
        else:
            d1[x]=d1[x]+1
    
    for y in t:
        if y not in d1:
            return False;
        else:
            if d1[y]==0:
                return False
            d1[y]=d1[y]-1
       
    return True;

#Return the top k most frequent words
def top_k_words(s, k): 
    # add your code here
    str1="!?',;."  
    output = {}
    list=[]
    for char in str1:
        s=s.replace(char,"")

    for word in s.split():
        if word not in output.keys():
            output[word] = 1
        else:
            output[word] += 1
    sorted_dict=sorted(output, key=lambda word: (-output[word], word))

    for key in sorted_dict:
        list.append(key)
    return list[:k]
   
#Calculate the Gradient descent
def gradient_descent(x, y, m, c, epochs, L=0.001):
    # add your code here
    n=len(x)
    dm=[]
    dc=[]
    y_pred=[]
    x1=[]
    for i in range(n):
        dm.append(0.0)
        dc.append(0.0)
        y_pred.append(0.0)
        x1.append(x[i][0])

    while(epochs!=0):
        for i in range(len(x1)):
            y_pred[i]=x1[i]*m+c
            dm[i]=x1[i]*(y_pred[i]-y[i])
            dc[i]=y_pred[i]-y[i]
        DM=sum(dm)/len(dm)
        DC=sum(dc)/len(dc)
        m=m-(L*DM)
        c=c-(L*DC)
        epochs-=1
    
    return m , c
#

if __name__ == "__main__": 
    
    # Test Question 1
    
    print("\nQ1")
    q1_test1 = 121
    q1_test2 = -121
    q1_test3 = 0
    q1_answer1 = is_palindrome(q1_test1)
    q1_answer2 = is_palindrome(q1_test2)
    q1_answer3 = is_palindrome(q1_test3)
    print(q1_answer1 )
    print("right answer: True")
    print("--------------")
    print(q1_answer2)
    print("right answer: False")
    print("--------------")
    print(q1_answer3)
    print("right answer: True")

    
    print("\nQ2")
    q2_test1_s = "anagram"
    q2_test1_t = "nagaram"
    q2_answer1 =  is_anagrams(q2_test1_s, q2_test1_t)
    print(q2_answer1)
    print("right answer: True")

    print("--------------")
    q2_test2_s = "python"
    q2_test2_t = "py"
    q2_answer2 =  is_anagrams(q2_test2_s, q2_test2_t)
    print(q2_answer2)
    print("right answer: False")
    print("--------------")

    # test question 3
    print("\nQ3")
    q3_test1_s = "   i love python, he    love coding python. the course is about python. "
    q3_test1_k = 2
    q3_answer = top_k_words(q3_test1_s, q3_test1_k)
    print(q3_answer)
    print("right: answer:")
    print("['python', 'love']")

    print ("\nQ4")

    x = x = [[0.18],[1.0],[0.92],[0.07],[0.85],[0.99],[0.87]]
    y = [109.85, 155.72, 137.66, 76.17, 139.75, 162.6, 151.77]
    m = 0
    c = 0
    q4_epochs200 = 200
    q4_epochs500 = 500
    q4_epochs1000 = 1000
    q4_epochs2000 = 2000
    q4_epochs3000 = 3000
    q4_answer1 = gradient_descent(x,y,m,c,q4_epochs200)
    q4_answer2 = gradient_descent(x,y,m,c,q4_epochs500)
    q4_answer3 = gradient_descent(x,y,m,c,q4_epochs1000)
    q4_answer4 = gradient_descent(x,y,m,c,q4_epochs2000)
    q4_answer5 = gradient_descent(x,y,m,c,q4_epochs3000)
    print(q4_answer1)
    print("right answer:")
    print("17.724810647355827, 22.97599012903927")
    print("--------------")
    print(q4_answer2)
    print("right answer:")
    print("35.97890301691016, 46.54235227399102")
    print("--------------")
    print(q4_answer3)
    print("right answer:")
    print("52.816639894324545, 68.05971340716786")
    print("--------------")
    print(q4_answer4)
    print("right answer:")
    print("64.56549666509812, 82.46678636085996")
    print("--------------")
    print(q4_answer5)
    print("right answer:")
    print("67.42648874428104, 85.32444456113602")
    print("--------------")
```
